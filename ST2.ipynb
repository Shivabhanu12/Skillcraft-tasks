```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/gender_submission.csv'
data = pd.read_csv(file_path)

# 1. Data Cleaning

# Check for missing values
missing_values = data.isnull().sum()

# Ensure 'Survived' is a categorical variable
data['Survived'] = data['Survived'].astype('category')

# 2. Exploratory Data Analysis (EDA)

# Check the distribution of survival
survival_counts = data['Survived'].value_counts()

# Plot the survival count
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=data, palette='Set1')
plt.title('Survival Count')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.ylabel('Number of Passengers')
plt.show()

# 3. Survival percentage
survival_percentage = survival_counts / len(data) * 100
print("Survival percentages:\n", survival_percentage)

# 4. Summary statistics
summary_stats = data.describe()

# 5. Visualizing survival rate
plt.figure(figsize=(8, 6))
data['Survived'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'skyblue'], labels=['Did Not Survive', 'Survived'], startangle=90)
plt.title('Survival Rate')
plt.ylabel('')
plt.show()

# 6. Passenger ID distribution (though not much meaning in this context)
plt.figure(figsize=(8, 6))
plt.hist(data['PassengerId'], bins=20, color='teal', alpha=0.7)
plt.title('Passenger ID Distribution')
plt.xlabel('Passenger ID')
plt.ylabel('Frequency')
plt.show()
```
